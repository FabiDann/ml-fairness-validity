```python
migra = pd.read_csv('data/migration_integration_regionen_daten.csv')
migra = migra[["NAME","INSGESAMT", "INSG_W", "AUSL_INSG", "AUSL_W"]]
migra
```

```python
f = lambda x: (x.replace(",",".")) 
krim = pd.read_csv("data/kriminalstatistik_kreise.csv", thousands=",", converters={"in Prozent": f})
krim["kreis"] = krim["Stadt-/Landkreis"] + " " + krim["Kreisart"]
krim = krim.loc[krim['Straftat'] == "Straftaten insgesamt"][["kreis","Fälle","Nichtdeutsche Tatverdächtige", "in Prozent"]]
krim
```

```python
lookup = pd.read_csv("data/join_kreise.csv")
agg = rki.groupby(['Landkreis', "Bundesland"]).sum()
joined = lookup.merge(agg, left_on='pk_rki', right_on='Landkreis').merge(krim, left_on="pk_krim", right_on="kreis").merge(migra, left_on="pk_migra", right_on="NAME")
joined.shape
joined.describe()
joined
```

```python
df = joined[["AnzahlFall", "in Prozent","Nichtdeutsche Tatverdächtige", "INSGESAMT", "AUSL_INSG", "AnzahlTodesfall"]]
X = df[:-20]
train = df[:-20]
test = df[-20:]
X_train = train[["AnzahlFall", "in Prozent","Nichtdeutsche Tatverdächtige", "INSGESAMT", "AUSL_INSG"]]
Y_train = train[["AnzahlTodesfall"]]
X_test = test[["AnzahlFall", "in Prozent","Nichtdeutsche Tatverdächtige", "INSGESAMT", "AUSL_INSG"]]
Y_test = test[["AnzahlTodesfall"]]

lreg = LinearRegression()
lreg.fit(X_train, Y_train)

Y_pred = lreg.predict(X_test)

err = Y_pred - Y_test


print(Y_pred.std())
print(Y_pred.mean())
print(err.std())
print(err.mean())
```
