```python
f = lambda x: (x.replace(",",".")) 
krim = pd.read_csv("data/kriminalstatistik_kreise.csv", thousands=",", converters={"in Prozent": f})
krim["kreis"] = krim["Stadt-/Landkreis"] + " " + krim["Kreisart"]
krim = krim.loc[krim['Straftat'] == "Straftaten insgesamt"][["kreis","Fälle","Nichtdeutsche Tatverdächtige", "in Prozent"]]

krim
```

```python
lookup = pd.read_csv("data/join_kreise.csv")
agg = rki.groupby(['Landkreis', "Bundesland"]).sum()
joined = lookup.merge(agg, left_on='pk_rki', right_on='Landkreis').merge(krim, left_on="pk_krim", right_on="kreis").merge(migra, left_on="pk_migra", right_on="NAME")
joined.shape
joined.describe()
```
